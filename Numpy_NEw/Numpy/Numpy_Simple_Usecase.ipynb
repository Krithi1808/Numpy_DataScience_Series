# Convert RGB to Black and White Image

```python
try:
    from scipy.datasets import face
except ImportError:  # Data was in scipy.misc prior to scipy v1.10
    from scipy.misc import face

img = face()
```

The image data. The returned array has shape

(M, N) for grayscale images.

(M, N, 3) for RGB images.

(M, N, 4) for RGBA images.

PNG images are returned as float arrays (0-1). All other formats are returned as int arrays, with a bit depth determined by the file's contents.

```python
img.shape

#RGB Image
```

First Index in 3rd Axis --> Red Pixel data

Second Index in 3rd Axis --> G Pixel Data

Third Index in 3rd Axis --> B Pixel Data

```python
img[:,:,0]
#Red Pixel Data
```

```python
img[:,:,1]
#Blue Pixel Data
```

## Lets convert pixel data to the range of (0,1)

```python
img_array=img/255
```

# we can assign each color channel to a separate matrix using the slice syntax:

```python
red_array = img_array[:, :, 0]
green_array = img_array[:, :, 1]
blue_array = img_array[:, :, 2]
```

# According to Colorimetry conversion from RGB to Grayscale can be achieved by formula


  **Y=0.2126R+0.7152G+0.0722B**

```python
Y=0.2126*red_array+0.7152*green_array+0.0722*blue_array


newimage=img_array@[0.2126,0.7152,0.0722]
Y
```
